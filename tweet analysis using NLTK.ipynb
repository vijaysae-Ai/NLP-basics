## Word Count

```python
df['word_count']=df['tweets'].apply(lambda x:len(str(x).split()));
```

```python
df.head()
```

```python
## Average words
def avg_word(x):
    word=x.split()
    word_count=0
    for i in word:
        word_count=word_count + len(i)
    return word_count/len(word)
```

```python
df['avg_word_count']=df['tweets'].apply(lambda x:avg_word(x))
```

## Preprocessing and Cleaning

```python
#Lower case 
df['tweets']=df['tweets'].apply(lambda x:x.lower())
```

```python
df.head()
```

## Count Email and remove Email

```python
import re
```

```python
x='hai grj idb  sae20@ ffbbf vfdjbhf vijay.sae20@gmail.com'
```

```python
re.findall(r'([a-zA-Z0-9+._-]+@[a-zA-Z0-9+._-]+\.[a-zA-Z0-9+._-]+)',x)
```

```python
df['email']=df['tweets'].apply(lambda x:re.findall(r'([a-zA-Z0-9+._-]+@[a-zA-Z0-9+._-]+\.[a-zA-Z0-9+._-]+)',x))
```

```python
df.head()
```

```python
df['email_count']=df['tweets'].apply(lambda x:len(re.findall(r'([a-zA-Z0-9+._-]+@[a-zA-Z0-9+._-]+\.[a-zA-Z0-9+._-]+)',x)))
```

```python
df.head()
```

```python
df['email_count'].value_counts()
```

```python
#remove
df['tweets']=df['tweets'].apply(lambda x:(re.sub(r'([a-zA-z0-9_.-]+@[a-zA-z0-9_.-]+\.[a-zA-z0-9_.-]+)',' ',x)))
```

```python
df['tweets']
```

## Count url and remove

```python
df['url_counts']=df['tweets'].apply(lambda x:len(re.findall(r'(http|https|ftp)://([\w_-])+(?:(?:[\w]))+([\w.,@#$%^&*:;])',x)))
```

```python
df.head()
```

```python
# remove RT
df['tweets']=df['tweets'].apply(lambda x: re.sub('RT' ,' ',x))
```

```python
## remove Punctuation 
df['tweets']=df['tweets'].apply(lambda x:re.sub('[^a-z A-z 0-9]+', '',x))
```

```python
df.head()
```

```python
df['tweets']=df['tweets'].apply(lambda x: " ".join(x.split()))
```

```python
df.head()
```
